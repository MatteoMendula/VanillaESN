| id            | Integer    |                      | no             |

## Training dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

training_data_location = 'occupancy_detection/datatraining.txt'
testing_data_location1 = 'occupancy_detection/datatest.txt'
testing_data_location2 = 'occupancy_detection/datatest2.txt'

# read data from txt file datatrainin.txt
training_data = pd.read_csv(training_data_location, sep=',', header=0)
training_data["Occupancy_highlight"] = training_data["Occupancy"].astype(int) * 1000
training_data.plot()
print("n sample", len(training_data))
print("columns", training_data.columns)
```

```python
testing_data1 = pd.read_csv(testing_data_location1, sep=',', header=0)
testing_data1["Occupancy_highlight"] = testing_data1["Occupancy"].astype(int) * 1000
testing_data1.plot()
print("n sample", len(testing_data1))
```

```python
testing_data2 = pd.read_csv(testing_data_location2, sep=',', header=0)
testing_data2["Occupancy_highlight"] = testing_data2["Occupancy"].astype(int) * 1000
testing_data2.plot()
print("n sample", len(testing_data2))
```

## Heatmaps

-------------------------------

### Training 

```python
# remove date column
training_data_heatmap = training_data.drop('date', axis=1)
# training_data_heatmap = training_data_heatmap.drop('Occupancy', axis=1)
training_data_heatmap = training_data_heatmap.drop('Occupancy_highlight', axis=1)
# normalize data in order to have all values between 0 and 1
training_data_heatmap = (training_data_heatmap - training_data_heatmap.min()) / (training_data_heatmap.max() - training_data_heatmap.min())
# plot heatmap for training data while keeping columns names
plt.matshow(training_data_heatmap.corr())
plt.xticks(range(training_data_heatmap.shape[1]), training_data_heatmap.columns, fontsize=10, rotation=90)
plt.yticks(range(training_data_heatmap.shape[1]), training_data_heatmap.columns, fontsize=10)
# print correlation values inside cells
for (i, j), val in np.ndenumerate(training_data_heatmap.corr()):
    plt.text(j, i, '{:0.1f}'.format(val), ha='center', va='center', fontsize=10, color='red')
plt.colorbar()

training_data_heatmap.plot()
```
